```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Health_insurance.xls")
data.head()
```

```python
categorical = [feature for feature in data.columns if data[feature].dtype == "O"]
```

```python
categorical
```

```python
one_hot_encoded_data = pd.get_dummies(data, columns=categorical)
one_hot_encoded_data.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(one_hot_encoded_data.corr(), cmap="YlGnBu", annot=True)
```

```python
one_hot_encoded_data.head()
```

```python
one_hot_encoded_data.to_csv("final.csv", index=False)
```
